```python
import hashlib
from sklearn.impute import SimpleImputer


# Bước 2: Loại bỏ các biến chất lượng thấp
threshold = len(df) * 0.5
df_cleaned = df.dropna(thresh=threshold, axis=1)

# Bước 3: Xử lý các giá trị bị thiếu
# Thay thế giá trị thiếu bằng trung vị 
df_cleaned.fillna(df.median(), inplace=True)

# Bước 4: Tạo chuỗi đại diện cho mỗi hàng bằng cách kết hợp tất cả các cột thành một chuỗi duy nhất
df_cleaned['combined_columns'] = df_cleaned.astype(str).agg(' '.join, axis=1)

# Bước 5: Tính hàm băm SHA256 cho từng hàng dựa trên chuỗi đã kết hợp
df_cleaned['hash'] = df_cleaned['combined_columns'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest())

# Bước 6: Loại bỏ các hàng có hàm băm SHA256 trùng lặp
df_unique = df_cleaned.drop_duplicates(subset=['hash'])

# Xóa các cột trung gian (nếu cần)
df_unique = df_unique.drop(columns=['combined_columns', 'hash'])

# Bước 7: Lưu dữ liệu đã loại bỏ trùng lặp vào file CSV mới
df_unique.to_csv('TUANDROMD_cleaned.csv', index=False)

print(f"Số lượng hàng sau khi loại bỏ các hàng trùng lặp: {df_unique.shape[0]}")
```

```python
import pandas as pd
import numpy as np

# Bước 1: Đọc dữ liệu từ tệp CSV
df = pd.read_csv('TUANDROMD_cleaned.csv')

# Bước 2: Tiền xử lý dữ liệu (nếu cần)

# Bước 3: Tính Entropy của biến mục tiêu
def entropy(target):
    value_counts = target.value_counts()
    probabilities = value_counts / len(target)
    return -np.sum(probabilities * np.log2(probabilities + 1e-9))  # Tránh log(0)

target_variable = df['Label']  
entropy_target = entropy(target_variable)

# Bước 4: Tính Conditional Entropy cho mỗi đặc trưng
def conditional_entropy(feature, target):
    feature_values = feature.unique()
    conditional_entropies = 0

    for value in feature_values:
        subset = target[feature == value]
        conditional_entropies += (len(subset) / len(target)) * entropy(subset)

    return conditional_entropies

# Bước 5: Tính Gain và Gain Ratio cho mỗi đặc trưng
gain_ratios = {}

for column in df.columns:
    if column != 'Label':  # Bỏ qua cột mục tiêu
        cond_entropy = conditional_entropy(df[column], target_variable)
        gain = entropy_target - cond_entropy

        # Tính giá trị nội tại của đặc trưng
        intrinsic_value = entropy(df[column])

        # Tính Gain Ratio
        gain_ratio = gain / (intrinsic_value + 1e-9)  # Tránh chia cho 0
        gain_ratios[column] = gain_ratio

# Bước 6: Lọc các đặc trưng có Gain-Ratio cao nhất cho mỗi loại
top_n = 20  # Số lượng đặc trưng cần chọn cho mỗi nhóm

# Tạo DataFrame từ gain_ratios
gain_ratios_df = pd.DataFrame(list(gain_ratios.items()), columns=['Feature', 'Gain-Ratio'])

# Lọc cho từng nhóm
top_permissions = gain_ratios_df[gain_ratios_df['Feature'].isin(permissions)].nlargest(top_n, 'Gain-Ratio')
top_api_calls = gain_ratios_df[gain_ratios_df['Feature'].isin(api_calls)].nlargest(top_n, 'Gain-Ratio')
top_intents = gain_ratios_df[gain_ratios_df['Feature'].isin(intents)].nlargest(top_n, 'Gain-Ratio')

# Bước 7: Tạo danh sách tất cả các đặc trưng đã chọn bao gồm cột label
selected_features = list(top_permissions['Feature']) + list(top_api_calls['Feature']) + list(top_intents['Feature']) + ['Label']

# Bước 8: Tạo DataFrame mới chỉ chứa các đặc trưng đã chọn
final_df = df[selected_features]

# Bước 9: Lưu DataFrame mới vào tệp CSV
final_df.to_csv('TUANDROMD_gain.csv', index=False)
 
# In ra thông báo hoàn tất
print("Dataset với các đặc trưng đã chọn đã được lưu vào 'TUANDROMD_gain.csv'")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Đọc dữ liệu từ tệp CSV
df = pd.read_csv('TUANDROMD_gain.csv')

# Kiểm định Chi-Squared cho từng loại đặc trưng
def chi_squared_test(feature, target):
    # Tạo bảng tần suất
    contingency_table = pd.crosstab(df[feature], df[target])
    
    # Thực hiện kiểm định Chi-Squared
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    return chi2, p

# Kiểm định Chi-Squared cho Permission
results_permissions = {}
for permission in top_permissions['Feature']:  # Duyệt qua danh sách top_permissions
    chi2, p_value = chi_squared_test(permission, 'Label')
    results_permissions[permission] = (chi2, p_value)

# Kiểm định Chi-Squared cho API calls
results_api_calls = {}
for api_call in top_api_calls['Feature']:  # Duyệt qua danh sách top_api_calls
    chi2, p_value = chi_squared_test(api_call, 'Label')
    results_api_calls[api_call] = (chi2, p_value)

# Kiểm định Chi-Squared cho Intents
results_intents = {}
for intent in top_intents['Feature']:  # Duyệt qua danh sách top_intents
    chi2, p_value = chi_squared_test(intent, 'Label')
    results_intents[intent] = (chi2, p_value)

# Kết quả
print("Kết quả kiểm định Chi-Squared cho Permissions:")
for permission, (chi2, p) in results_permissions.items():
    print(f"{permission}: Chi2 = {chi2:.2f}, p-value = {p:.4f}")

print("\nKết quả kiểm định Chi-Squared cho API calls:")
for api_call, (chi2, p) in results_api_calls.items():
    print(f"{api_call}: Chi2 = {chi2:.2f}, p-value = {p:.4f}")

print("\nKết quả kiểm định Chi-Squared cho Intents:")
for intent, (chi2, p) in results_intents.items():
    print(f"{intent}: Chi2 = {chi2:.2f}, p-value = {p:.4f}")

# Lọc các đặc trưng có p-value nhỏ hơn 0.05
significant_permissions = {k: v for k, v in results_permissions.items() if v[1] < 0.05}
significant_api_calls = {k: v for k, v in results_api_calls.items() if v[1] < 0.05}
significant_intents = {k: v for k, v in results_intents.items() if v[1] < 0.05}

# Bước 7: Tạo danh sách tất cả các đặc trưng đã chọn bao gồm cột label
final_features = list(significant_permissions.keys()) + list(significant_api_calls.keys()) + list(significant_intents.keys()) + ['Label']

# Bước 8: Tạo DataFrame mới chỉ chứa các đặc trưng đã chọn
final_df = df[final_features]

# Bước 9: Lưu DataFrame mới vào tệp CSV
final_df.to_csv('TUANDROMD_chi_square.csv', index=False)

# In ra thông báo hoàn tất
print("Dataset với các đặc trưng đã chọn đã được lưu vào 'TUANDROMD_chi_square.csv'")
```
